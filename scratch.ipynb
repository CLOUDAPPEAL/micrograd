```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 0.5 * x**2 + 3 * x + 4
```

```python
f(3.0)
```

```python
xs = np.arange(-10, 10, 0.5)
ys = f(xs)
print(ys)
plt.plot(xs, ys)
```

```python
x = -1.0
h = 0.0001
print(f(x))
print(f(x + h))
(f(x+h) - f(x)) / h
```

```python
a = 1.0
b = 3.0
c = -5.0
d = a * b + c
print(d)
```

```python
a = 1.0
b = 3.0
c = -5.0
h = 0.0001

d1 = a * b + c
a += h 
d2 = a * b + c

print("d1:", d1)
print("d2:", d2)
print("d2 - d1:", d2 - d1)
print("slope:", (d2 - d1) / h)
```
